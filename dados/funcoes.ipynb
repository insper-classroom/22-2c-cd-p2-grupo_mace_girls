```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import re 
```

```python
dados = pd.read_csv('../dados/titles.csv')
dados.tmdb_popularity.describe()
```

```python
def genero(csv):
    dados = csv
    lista_limpa = []
    for i in range(0,len(dados)):
        lista = dados['genres'].iloc[i].split(',')
        for x in lista:
            p = ''
            for l in x:
                if l != '[' and l !=']':
                    p += l
            lista_limpa.append(p)
                
    return(lista_limpa)

genero(dados)
```

```python
print('Fizemos a função gênero para conseguir analisar, dado a nossa target = popularidade dos filmes,\nqual a relação entre o gênero dele e a sua popularidade')
def generos(genero, csv):
    dados = csv
    dados[genero] = dados['genres']
    for i in range(0,len(dados)):
            lista = dados['genres'].iloc[i]
            if genero in lista:
                dados[genero][i] = 1
            else:
                dados[genero][i] = 0
    return(dados)
print('essa é a base de dados com alguns gêneros aplicados pela função')
dados = generos('comedy', pd.read_csv('../dados/titles.csv'))
dados = generos('romance', dados)
dados = generos('drama', dados)
dados = generos('action', dados)
dados = generos('fantasy', dados)
dados = generos('horror', dados)
dados.head(5)
```

```python
print('Para saber quais os gêneros mais comuns nos filmes e eliminar os menos relevantes, fizemos essa função para analisar gênero e popularidade, que recebe a base de dados e retorna uma lista com todos os gêneros')
filmes= dados.loc[(dados.type == 'MOVIE'), :]
def comum(dados):
    dados = dados.genres
    generos = []
    for i in range(0,len(dados)):
        lista = dados[i]
        for x in lista:
            if x not in generos:
                generos.append(x)
        i+=1
    return(generos)
gen_filmes = comum(dados) 
print(gen_filmes)
```

```python
def scatter (v1, v2):
    plt.scatter(v1, v2, alpha=0.4)
    plt.title('Gráfico de dispersão entre popularidade e quantidade de votos')
    plt.xlabel('quantidade de votos')
    plt.ylabel('popularidade')
    plt.subplots_adjust(top=1)
    print('Correlação entre tmdb_popularity e imdb_votes é {0:.2f}'.format(v2.corr(v1)))
    return(plt.show())
```

```python
#função para achar o limite superior e limite inferior das variáveis quantitativas
filmes= dados.loc[(dados.type == 'MOVIE'), :]

def valores_extremos(dados, vari):
    dados= dados[vari]
    Qter= dados.quantile(0.75)
    Qpri= dados.quantile(0.25)
    iq= Qter - Qpri
    LS = Qter + 1.5*iq
    LI= Qpri - 1.5*iq
    return(dados.loc[(dados>LI)&(dados<LS)])

valores_extremos(filmes, 'tmdb_popularity')
```
